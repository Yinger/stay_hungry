```python

import matplotlib.pyplot as plt
import pandas as pd
import requests

# 

# 
periods = '3600'

# Httpbtc
resp = requests.get('https://api.cryptowat.ch/markets/gemini/btcusd/ohlc', 
  params={
    'periods': periods
  })
data = resp.json()

# pandas data frame
df = pd.DataFrame(
  data['result'][periods], 
  columns=[
    'CloseTime', # 
    'OpenPrice', # 
    'HighPrice', # 
    'LowPrice', # 
    'ClosePrice', # 
    'Volume', # 
    'NA'])

# DataFrame
print(df.head())

# btc
df['ClosePrice'].plot(figsize=(14, 7)) # 

# ax = df["ClosePrice"].plot(figsize = (14, 7))
# 	fig = ax.get_figure()
# 	fig.savefig("closeprice.png")
```
