## 監督學習（Supervised Learning）
監督學習是機器學習的一種方法，指從訓練數據（輸入和預期輸出）中學到一個模型（函數），并根據模型可以推斷新實例的方法。
函數的輸出通常爲一個連續值（回歸分析）或類別標簽（分類）。
### 監督學習典型算法

* 邏輯回歸（Logistic Regression）

* 隨機森林（Random Forest）
* 最近鄰算法（k-NN）
* 樸素貝葉斯（Naive Bayes）
* 支持向量機（SVM）
* 感知器（Perceptron）
* 深度神經網絡（DNN）

## 綫性回歸（Linear Regression）
在統計學中，綫性回歸是利用稱爲綫性回歸方程的 __最小二乘函數__ 對 __一個__或 __多個__ 自變量和因變量之間關係進行建模的一種回歸分析。這種函數是一個或多個稱爲回歸係數的模型參數的綫性組合

### 單變量綫性回歸
* 理想函數

> y = b + wx

* 假設函數

>$h_\theta(x) = \theta_0 + \theta_1x_1 = \theta^Tx$

>$h_\theta(x) = \theta_0x_0 + \theta_1x_1 = \theta^Tx$     $(x_0 = 1)$ 

* 損失值（誤差）

>$loss = y - h_\theta(x)$

爲了從一組樣本 $ \left( x^{(i)} , y^{(i)} \right) $ (其中i = 1,2,...,n)之中估計最合適（誤差最小）的 $\theta_0$ 和 $\theta_1$ ，通常采用最小二乘法，其優化目標為最小化殘差平方和

> $J(\theta) = \frac{1}{n} \sum_{i}^n (h_\theta(x^{(i)}) - y^{(i)}) ^2$

> $J(\theta) = \frac{1}{2n} \sum_{i}^n (h_\theta(x^{(i)}) - y^{(i)}) ^2$

* 梯度下降

模型參數 $\theta_0x_0$ 和 $\theta_1x_1$在優化目標函數的每輪迭代中，按如下表達式更新

> $\theta_j := \theta_j-\alpha\frac{\partial}{\partial\theta_j}J(\theta) $

代入$J(\theta)$求導，結果如下

> $\theta_j := \theta_j - \alpha\frac{\partial}{\partial\theta_j} \frac{1}{2n} \sum_{i}^n (h_\theta(x^{(i)}) - y^{(i)}) ^2 $

> $\theta_j := \theta_j - \alpha \frac{1}{2n} \sum_{i}^n 2(h_\theta(x^{(i)}) - y^{(i)}) (x_j^{(i)}) $

> $\theta_j := \theta_j - \alpha \frac{1}{n} \sum_{i}^n (h_\theta(x^{(i)}) - y^{(i)}) (x_j^{(i)}) $

### 多變量綫性回歸
* 理想函數

> $y = w_0 + w_1x_1 + w_2x_2$

> $ \overrightarrow {y} = W ^T X $

* 假設函數

>$h_\theta(X) = \theta^TX = \theta_0 + \theta_1x_1 + \theta_2x_2$

* 損失值（誤差）

>$loss = \overrightarrow {y} - h_\theta(X)$

爲了從一組樣本 $ \left( x^{(i)}_1 , x^{(i)}_2 , y^{(i)} \right) $ (其中i = 1,2,...,n) 之中估計最合適（誤差最小）的模型參數 
$ \theta_j( \theta_0 , \theta_1 , \theta_2 ) $，目標函數如下：

> $J(\theta) = \frac{1}{2n}(h_\theta(X) - \overrightarrow {y})^T (h_\theta(X) - \overrightarrow {y})$

> $J(\theta) = \frac{1}{2n}(X\theta - \overrightarrow {y})^T (X\theta - \overrightarrow {y})$

> $X = \left[ (x^{(1)})^T ... (x^{(n)})^T \right]$

> $\overrightarrow {y} = \left[ y^{(1)} ... y^{(n)} \right]$

* 梯度下降

與單變量綫性回歸不同，多變量綫性回歸的最大問題是容易出現 __權值越界__，多變量現象回歸有多個特徵值，每一個特徵值的取值範圍相差各異，最大與最小往往相差幾十甚至幾百個個數量級，在訓練時由於不同特徵值的權重對神經網絡的貢獻不同，就容易造成權重的差距特別大，進而導致權重越界，爲此需要對每個特徵值進行 __預處理__ (特徵縮放 Scaling)，最常用的預處理方式是 __歸一化__，廣義的歸一化既是將特徵值的大小限制在 0～1 之間，常用的歸一化方式有除以數據最值，以及設定均值方差等標準等

> $x' = \frac {x_i - mean(x)} {std(x)}$

> $x' = \frac {x_i} {max(x) - min(x)}$

```python
import numpy as np
# 向量属于矩阵范畴，至少两个维度

# 矩阵
matrix_value = [[1, 2, 3], [1.1, 2.2, 3.3]]
print('matrix_value type:',type(matrix_value))
print('matrix_value type:',matrix_value)

# transform list to matrix
matrix_np = np.array(matrix_value)
print('matrix_np type :', type(matrix_np))
print('matrix_np shape:', matrix_np.shape)
print('matrix_np value:\n', matrix_np)
```

```python
# 行向量
vector_row = np.array([[1, 2, 3]])
print(vector_row, 'shape=', vector_row.shape)
```

```python
# 列向量
vector_column = np.array([[1], [2], [3]])
print(vector_column, 'shape=', vector_column.shape)
```

```python
# 行列转置，使用 numpy.T 进行转置，对任意维度矩阵适用
print(vector_row, 'shape=', vector_row.shape, '\n',
      '↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓\n',
      vector_row.T, 'Tshape=', vector_row.T.shape)

print('\n')

matrix = np.array([[1, 2, 3], [4, 5, 6]])
print('raw matrix:\n', matrix,
      '\nraw matrix shape:', matrix.shape,
     '\nTmatrix:\n', matrix.T,
     '\nTmatrix shape:', matrix.T.shape)
```

```python
# 矩阵运算仅讨论矩阵乘除，加减太易舍去
# 点乘有两种方式 ，* & numpy.multiply
matrix_1 = np.array([[1, 2, 3], [4, 5, 6]])
matrix_2 = np.array([[1, 2, 3], [4, 5, 6]])
matrix_result_1 = matrix_1 * matrix_2
matrix_result_2 = np.multiply(matrix_1, matrix_2)
if (matrix_result_1==matrix_result_2).all():
    print('matrix_result_1 is equal to matrix_result_2')
print('matrix_1 multiply matrix_2 is:\n', matrix_result_1)

print('\n')

# 叉乘只有 matmul 一种计算方式，是线性代数，机器学习，图像处理较为默认的乘法
matrix_1 = np.array([[1, 2, 3], [4, 5, 6]])
matrix_2 = np.array([[1, 2], [3, 4], [5, 6]])
matrix_result_3 = np.matmul(matrix_1, matrix_2)
print('matrix_1 matmul matrix_2 is:\n', matrix_result_3)
```

```python
# numpy 拓展部分，判断矩阵是否相等
# any() 用于每一个元素相对应判断，存在元素相等，则相等
# all() 用于判断矩阵是否全等，全部元素相等则相等
# 判断两个矩阵相等时，最好需要维度相同，维度不同会报严重警告(elementwise comparison failed; this will raise an error in the future)，甚至引起程序崩溃

#矩阵比较返回元素为 bool 值的矩阵
matrix_1 = np.array([[1, 2, 3], [4, 5, 6]])
matrix_2 = np.array([[1, 3, 2], [4, 5, 7]])
bool_matrix = matrix_1==matrix_2
print('bool_matrix:\n', bool_matrix)

print('matrix_1 is equal to matrix_2 in some positions')
print('bool_matrix.any():', bool_matrix.any())
print('matrix_1 is not equal to matrix_2 in all positions')
print('bool_matrix.all():', bool_matrix.all())
```

# TensorFLow 2.x 實現

## 導入必要的包

```python
import tensorflow as tf
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
from sklearn.preprocessing import scale
import os

tf.__version__
```

## 讀取數據

```python
# 读取数据并转换为 numpy array 格式
data_frame = pd.read_csv('data1.csv', names=['square', 'bedrooms', 'price'])
data_frame.head()
```

```python
# 數據可視化

fig = plt.figure()

# 创建一个 Axes3D object
ax = plt.axes(projection='3d')

# 设置 3 个坐标轴的名称
ax.set_xlabel('square')
ax.set_ylabel('bedrooms')
ax.set_zlabel('price')

# 绘制 3D 散点图
ax.scatter3D(data_frame['square'], data_frame['bedrooms'], data_frame['price'], c=data_frame['price'], cmap='Greens')
```

## 数据规范化
* 使用 sklearn 的 preprocessing 下的 scale 函数进行归一化

> $x' = \frac {x_i - mean(x)} {std(x)}$

```python
def normalize_feature(df):
    return df.apply(lambda column: (column - column.mean()) / column.std())

data_frame = normalize_feature(data_frame)

print(data_frame.head())
X_data = np.array(data_frame[data_frame.columns[0:3]])
y_data = np.array(data_frame[data_frame.columns[-1]]).reshape(len(data_frame), 1)
```

```python
ax = plt.axes(projection='3d')
ax.set_xlabel('square')
ax.set_ylabel('bedrooms')
ax.set_zlabel('price')
ax.scatter3D(data_frame['square'], data_frame['bedrooms'], data_frame['price'], c=data_frame['price'], cmap='Reds')
```

```python
data_frame.info()
```

## 構建模型

```python
# 使用 tf.keras 管理 Sequential 模型
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(1,input_shape = (3,)))
model.summary()
```

## 訓練

```python
# 定義優化器，損失函數，關心指標
model.compile(optimizer = 'sgd', # 優化器
              loss = 'mean_squared_error', # loss function
              metrics=['mae'] # 關心的指標
)

# 訓練
history = model.fit(X_data, y_data, epochs = 50) # epochs:迭代的輪數

# hist = pd.DataFrame(history.history)
# hist['epoch'] = history.epoch
```

## 評估

```python
model.evaluate(X_data, y_data) # 本來應該用沒有被訓練過的數據做評估，這裏數據集比較少，暫時先掌握evaluate用法
```

## 保存模型

```python
model.save("house_price_model")
```

## 加載模型

```python
sm = tf.keras.models.load_model("house_price_model")
sm.evaluate(X_data, y_data)
```
